```python
def print_board(status):
    """
    recives a monodimensional list of values and prints the board from those.
    """
        
    dividing_line = "  --- --- ---  "
    
    print(dividing_line)
    print(" | {} | {} | {} |".format(*status[0:3]))
    print(dividing_line)
    print(" | {} | {} | {} |".format(*status[3:6]))
    print(dividing_line)
    print(" | {} | {} | {} |".format(*status[6:9]))
    print(dividing_line)
```

```python
def clear_board():
    """
    returns a clear board
    """
    
    game = [[" ", " ", " "],
            [" ", " ", " "],
            [" ", " ", " "]]
    return game
```

```python
def check_winner(board):
    """
    check is somebody won. "X" is player 1. "O" is player 2
    """
    
    # check all the rows
    for i in board:
        if i == ["X", "X", "X"]:

            return 1
        elif i == ["O", "O", "O"]:
            #print("player 2 won")
            return 2
            
    # check all the columns
    for j in range(3):
        if (board[0][j] == board[1][j] == board[2][j] == "X"):
            return 1
        elif (board[0][j] == board[1][j] == board[2][j] == "O"):
            return 2
        
    # check first diagonal
    if (board[0][0] == board[1][1] == board[2][2] == "X"):
        return 1
    elif (board[0][0] == board[1][1] == board[2][2] == "O"):
        return 2
    
    # check second diagonal
    if (board[2][0] == board[1][1] == board[0][2] == "X"):
        return 1
    elif (board[2][0] == board[1][1] == board[0][2] == "O"):
        return 2    

    # otherwise no winner
    else:
        return 0
```

```python
def main():
    
    game = clear_board()
    playing = True

    while playing:

        unrolled = [elem for row in game for elem in row]
        counter = unrolled.count("X") + unrolled.count("O")

        print_board(unrolled)

        resp = check_winner(game)

        if resp == 1:
            print("Congratulations Player 1, you won!")
            playing = False
            break
        elif resp == 2:
            print("Congratulations Player 2, you won!")
            playing = False
            break
        elif counter == 9:
            print("Even Game!")
            playing = False
            break
            

        try:
            command = str(input("Insert comma separated coordinates. Q to exit: "))

            if command in ("Q", "q", "exit", "end", "finish", "bye"):
                print("Thank for playing, see you soon")
                break
            else:
                x,y = command.split(",")
                x, y = int(x) - 1, int(y) -1

                if game[x][y] == " ":

                    if counter%2 == 0:
                        game[x][y] = "X"
                    else:
                        game[x][y] = "O"
                else:
                    print("This spot is already taken")

        except IndexError:
            print("The board is not big enough!")
        except ValueError:
            print("Retry")
    
```

```python
if __name__ == "__main__":
    main()
```
